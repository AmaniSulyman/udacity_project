**creatre the needed function**

```python
# creat naming fancion
def lables_name( x , y = None):
    plt.xlabel(x)
    if (y == None):
        plt.title(f' {x} Distribution',fontsize = 12, weight = "bold");
        plt.ylabel('scale',fontsize = 12, weight = "bold")
    else:
        plt.title(f' {x} and {y} ',fontsize = 12, weight = "bold")
        plt.ylabel(y, fontsize = 12, weight = "bold")
```

```python
# create function for histogram plotting
def Plot_hist(df,xVar,a,y = None):
    
    bins1 = np.arange(df[xVar].min(), df[xVar].max()+a, a)
    plt.figure(figsize=[14, 6])
    df[xVar].hist(data = df, bins = bins1,color=['y'])
    lables_name(xVar)
```

```python
# create function for sctter plotting
def plot_sctter(df,x,y):
    
    plt.figure(figsize = [14, 6])
    plt.scatter(df[x],df[y],color=['y'])
    lables_name(x,y)
    plt.title(f' state of {x} response to {y}')
```
